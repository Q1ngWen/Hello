```python
# import cell
import os
import numpy as np
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset into dataframe
spotify_data = pd.read_csv('https://www.inf.ed.ac.uk/teaching/courses/fds/data/project-2021-2022/spotify/data.csv.zip')
spotify_data
```

```python
# -------- initial data exploration --------
column_names = spotify_data.columns
column_names
```

```python
# a dictionary of initial unique values in each column (column name: unique values)
columns_unique_dict = {}
for name in column_names:
    columns_unique_dict[name] = spotify_data[name].unique()
columns_unique_dict
```

```python
# transform the region codes to all caps
spotify_data["Region"] = spotify_data["Region"].transform(lambda code: code.upper())
```

```python
# -------- data cleaning & processing --------
# remove the URL column - no use for it, excessive data
cleaned_data = spotify_data.copy()

# since we have a question to investigate which day of the week and month there are more streams, we add "Day Of Week" and "Day Of Month"
cleaned_data["Day Of Month"] = cleaned_data["Date"].transform(lambda date: int(date[-2:]))
cleaned_data["Day Of Week"] = cleaned_data["Date"].transform(lambda date: datetime.strptime(date,'%Y-%m-%d').weekday())
```

```python
# check for NaN values
for column in cleaned_data.columns:  # print out number of NaN values in each column: identify where the NaNs are
    print(column, cleaned_data[column].isnull().sum())

# drop na values in track names, as we definitely cannot predict those from the other values
cleaned_data = cleaned_data.dropna(axis=0, subset=["Artist"])

# print out number of NaN values in each column again
for column in cleaned_data.columns:
    print(column, cleaned_data[column].isnull().sum())
# Note: result is 0 - no more NaNs left!
# Cleaned 67 rows

# delete the "URL" column from our DataFrame
cleaned_data = cleaned_data.drop(columns="URL")
```

```python
cleaned_data
```

```python
regionCode_data = pd.read_csv("regioncodes_to_names.csv")
regionCode_data = regionCode_data[["name","alpha-2"]].copy()
regionCode_data
```

```python
regionName_dict = dict(zip(regionCode_data['alpha-2'], regionCode_data['name']))
regionName_dict['GLOBAL'] = 'Global'
regionName_dict["GB"] = "United Kingdom"
regionName_dict
```

```python
cleaned_data["Region Name"] = cleaned_data["Region"].transform(lambda code: regionName_dict[code])
```

```python
cleaned_data
```

```python
# add region names to our dictionary of unique column values
columns_unique_dict["Region Name"] = cleaned_data["Region Name"].unique()
columns_unique_dict
```

```python
fig_dayOfWeek, ax_dayOfWeek = plt.subplots(18,3)
fig_dayOfWeek.set_size_inches(10,55)

count = 0
for name in columns_unique_dict["Region Name"]:
    i = count // 3
    j = count % 3
    count += 1
    this_df = cleaned_data[cleaned_data["Region Name"] == name]
    ax_dayOfWeek[i][j].set_title(name)
    sns.boxplot(x="Day Of Week", y="Streams", data=this_df, ax=ax_dayOfWeek[i][j])
    ax_dayOfWeek[i][j].set_yscale('log')
fig_dayOfWeek.tight_layout()
```

```python
fig_dayOfWeek.savefig("streams_by_region_day.png")
```
